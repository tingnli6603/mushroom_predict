```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#選擇圖形樣式
%matplotlib inline
plt.style.use('fivethirtyeight')

import warnings
#忽略警告，為了排版整齊
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import LabelEncoder
```

## 探索式資料分析

### 基本資料

自變數:22個，皆為類別變數；因變數:1個，為類別變數。__是為類別變數的分類目的__

資料筆數: 8416 筆

```python
data_path = "./data/mushroom_dataset.csv"
mushroom_df = pd.read_csv(data_path)

mushroom_df
```

### 切分資料集

將訓練與測試資料切割，訓練80%、測試20%

```python
train_df, test_df = train_test_split(mushroom_df, test_size=0.2)

print(train_df.shape, test_df.shape)
```

### 遺漏值觀察

訓練資料與測試資料皆無遺漏值

```python
print(train_df.isnull().sum())
print(test_df.isnull().sum())
```

### 繪製長條圖

```python
#繪製每個類別的長條圖
columns = train_df.columns
print(columns)

for column_name in columns:
    print(column_name)
    
    print(train_df[column_name].value_counts())

    x_class = train_df[column_name].value_counts().keys()
    y_num = train_df[column_name].value_counts().values

    plt.bar(x_class, y_num)
    plt.show()
    
    plt.pie(y_num , labels = x_class,autopct='%1.1f%%')
    plt.show()
```

```python
#繪製每個類別與該類別與目標變數的長條圖
for column_name in columns:
    print("COLUMN NAME:", column_name)


    x_class = train_df[column_name].value_counts().keys()
    print("CLASS:", x_class)

    y_num = train_df[column_name].value_counts().values
    print("ALL:", y_num)

    y_edible_num = train_df[train_df['mushroom']=='EDIBLE'][column_name].value_counts()
    #某些標籤在不同的目標變數可能沒有值
    y_edible_list = []
    for class_name in x_class:
        if(class_name in y_edible_num.keys()):
            y_edible_list.append(y_edible_num[class_name])
        else:
            y_edible_list.append(0)

    print("EDIBLW:", y_edible_list)


    y_poisonous_num = train_df[train_df['mushroom']=='POISONOUS'][column_name].value_counts()
    y_poisonous_list = []
    for class_name in x_class:
        if(class_name in y_poisonous_num.keys()):
            y_poisonous_list.append(y_poisonous_num[class_name])
        else:
            y_poisonous_list.append(0)

    print("POISONOUS:", y_poisonous_list)

    plt.bar(x_class, y_edible_list, label = 'edible', align = "edge", width = 0.35)
    plt.bar(x_class, y_poisonous_list, label = 'poisonous', align = "edge", width = -0.35)
    plt.legend()

    plt.show()
```
